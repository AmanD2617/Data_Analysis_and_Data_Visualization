# Task 9: Plot a percentage segment bar graph between Education_Level and Attrition_Flag of the customers

```python
data = df.groupby('Education_Level')[['Attrition_Flag']].apply(lambda x: x.apply(pd.value_counts)).unstack().fillna(0)
data_percent = data.apply(lambda x: x / x.sum() * 100, axis=1)
data_percent.plot(kind='bar', stacked=True)
plt.xlabel('Education_Level')
plt.ylabel('Percentage')
plt.title('Stacked Bar Graph between Education_Level and Attrition_Flag of the customers')
plt.legend()
plt.show()
```

# Task 10:	Plot a percentage segment bar graph between Income_Category and Attrition_Flag of the customers

```python
data = df.groupby('Income_Category')[['Attrition_Flag']].apply(lambda x: x.apply(pd.value_counts)).unstack().fillna(0)
data_percent = data.apply(lambda x: x / x.sum() * 100, axis=1)
data_percent.plot(kind='bar', stacked=True)
plt.xlabel('Income_Category')
plt.ylabel('Percentage')
plt.title('Stacked Bar Graph between Income_Category and Attrition_Flag of the customers')
plt.legend()
plt.show()
```
